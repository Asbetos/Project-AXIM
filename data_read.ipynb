```python
import pandas as pd
import json
```

```python
with open(r"./all_schools.json", "r") as f:
    data = json.load(f)

df = pd.json_normalize(data, sep=".")

# The countries_most_represented column contains lists — convert to string to keep it readable
if "student_body_all_programs.countries_most_represented" in df.columns:
    df["student_body_all_programs.countries_most_represented"] = df["student_body_all_programs.countries_most_represented"].apply(str)

print(df.shape)
print(df.columns.tolist())

df.head()
```

```python
df.to_csv("usnews_schools.csv", index=False)
```
